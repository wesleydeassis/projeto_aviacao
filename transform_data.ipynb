```python
import pandas as pd
```

```python
#importar dados
# dados extraidos do site https://dados.gov.br/dados/conjuntos-dados/ocorrencias-aeronauticas-da-aviacao-civil-brasileira

df_ocorrencia = pd.read_csv(r"ocorrencia.csv", sep=",", encoding="utf-8")
df_aeronave = pd.read_csv(r"aeronave.csv", sep=";", encoding="ISO-8859-1")

```

```python
df_aeronave_new = df_aeronave[["codigo_ocorrencia2", "aeronave_matricula", "aeronave_voo_origem","aeronave_voo_destino", "aeronave_tipo_operacao", "aeronave_fatalidades_total"]]
df_aeronave_new
```

```python

#fazer merge dos dataframes
df_aeronave_new = df_aeronave_new[df_aeronave_new['aeronave_tipo_operacao']=='REGULAR']
df_aeronave_ocorrencia = pd.merge(df_aeronave_new, df_ocorrencia, left_on='codigo_ocorrencia2', right_on='codigo_ocorrencia', how='inner')
df_aeronave_ocorrencia


```

```python
#selecionar colunas de interesse e exportar como excel
df_aeronave_ocorrencia = df_aeronave_ocorrencia[["ocorrencia_dia","codigo_ocorrencia2","ocorrencia_uf","ocorrencia_classificacao", "aeronave_matricula", "aeronave_voo_origem","aeronave_voo_destino", "aeronave_tipo_operacao", "aeronave_fatalidades_total"]]

caminho_arquivo = r"C:\seucaminho\analise_aviacao\aeronaves_ocorrencias.xlsx"

df_aeronave_ocorrencia.to_excel(caminho_arquivo, index= False, sheet_name='Aeronaves_Ocorrências')
```
