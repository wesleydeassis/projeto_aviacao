```python
import pandas as pd
```

```python
#importar dados
# dados extraidos do site https://dados.gov.br/dados/conjuntos-dados/ocorrencias-aeronauticas-da-aviacao-civil-brasileira

df_ocorrencia = pd.read_csv(r"ocorrencia.csv", sep=",", encoding="utf-8")

df_aeronave = pd.read_csv(r"aeronave_v2.csv", sep=";", encoding="ISO-8859-1")

```

```python
#gerando novo dataframe com as colunas de interesse
df_aeronave_new = df_aeronave[["codigo_ocorrencia2", "aeronave_matricula", "aeronave_pais_registro", "aeronave_voo_origem","aeronave_voo_destino", "aeronave_tipo_operacao", "aeronave_fatalidades_total"]]
df_aeronave_new.head()
```

```python
#validando valores duplicados e nulos
df_aeronave_new.duplicated().sum()


```

```python
#validando valores duplicados e nulos
df_aeronave_new[df_aeronave_new.isnull().any(axis=1)]
```

```python
df_aeronave_new.dropna(inplace=True)
```

```python
#fazer merge dos dataframes

df_aeronave_ocorrencia = pd.merge(df_aeronave_new, df_ocorrencia, left_on='codigo_ocorrencia2', right_on='codigo_ocorrencia', how='inner')
```
